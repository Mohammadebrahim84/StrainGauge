# What is a Strain Gauge and How Do They Work?       
Strain is a dimensionless measurement that is a ratio of the change in length to the,
original length of an object. Therefore, a positive strain is the result of stretching
a material and negative strain is the result of compression. Stress is a measurement of
the force applied divided by the initial cross-sectional area of an object, or the internal,
resisting capacity of an object.
Each strain gauge is composed of a metal foil insulated by a flexible substrate, as
shown in the figure under. The two leads pass a current through the gauge, and as the,
surface of the object being measured stretches or contracts, the change in resistance is
measured. This change in resistance is proportional to the change in length on the,
surface of the object being tested, as shown in the equation below.  In simple terms: This formula states that the change in resistance of a Strain Gauge is directly proportional to the strain of the object to which it is attached.


$$ \frac{\Delta R}{R} = k \cdot \varepsilon $$  

Where:  
- \(R\) = Resistance (**Ω**)  
- \(ΔR\) = Change in resistance (**Ω**)  
- \(k\) = Gauge Factor  
- \(ε\) = Strain

# Strain Gauge Applications:            
The fields of civil engineering and geotechnical monitoring regularly use strain gauges
to detect failures in structures like bridges, buildings, and much more. These structures
require constant monitoring because any significant deformation could lead to injuries
or death. These gauges are commonly used because they have high precision, function
well at long distances from the test object, and require minimal effort to set up and
maintain over long periods of time.

Testing in the field is frequently quite different from laboratory testing under ideal
conditions. One of the reasons strain gauges are highly valued is the fact that they can
be used in harsh environments, yielding repeatable results with high precision. 
When an engineer tests objects with irregular shapes in harsh environments with difficult-to-access configurations,
a specialized device like a strain gauge is often needed. 
For example, aerospace applications use millions of strain gauges to verify results from
CAD (Computer Aided Design) and FEA (Finite Element Analysis) simulations.
These tests are often conducted in dynamic conditions to display an accurate representation of how different forces affect aircraft.

# Working Principle of Strain Gauge                  
Their working principle is based on piezo-resistive effect. The piezo-resistive effect states that if
a metal conductor is stretched or compressed,its length and diameter changes. Therefore its resistance also gets changed.
Similarly the resistance of a metal conductor changes when it is strained.

$$
R = \frac{\rho L}{A} = \frac{\rho L}{\left(\frac{\pi}{4} D^2\right)}
$$

$$
R = \frac{\rho L}{A} = \frac{\rho L}{\left(\frac{\pi}{4} D^2\right)}
$$


Where,

ρ = Resistivity of the wire in Ω – m

If this wire of elastic material (Strain Gauge) is subjected to tension, its length increases and diameter reduces.
</br>
</br>
![My Image](./images/img7.png)
</br>
Let the change in length = ΔL and change in diameter = ΔD. To find the change in resistance (ΔR) due to applied force we use Poisson’s ratio.

</br>

Using Poisson ratio (μ) = $$
\frac{\frac{\Delta D}{D}}{\frac{\Delta L}{L}}
$$


But,

$$\frac{\Delta L}{L} = \sigma = \text{Strain}$$
Therefore,

$$
\Delta D = \mu \cdot \sigma \cdot D
$$

Thus, the change in resistance is given by:

$$
\Delta R = R \sigma (1 + 2\mu)
$$

From the above equation, we can conclude that the resistance of the wire increases due to the applied force. The gauge factor (G) of a strain gauge is defined as the unit change in resistance per unit change in length. Therefore,

$$
G = \frac{\frac{\Delta R}{R}}{\frac{\Delta L}{L}}
$$


(Ranges from 1.5 to 1.7)






### Question1

You are working on a project in civil engineering and need to measure the strain on a metal rod subjected to tensile force. The metal rod has an initial resistance of 150 Ω and is equipped with a strain gauge. When the rod is stretched due to the applied force, the resistance changes by 1.2 Ω. The length of the rod before stretching is 2 meters, and after stretching, it changes by 0.002 meters. The gauge factor (G) for the strain gauge is 2.5.

Calculate the strain (ε) on the rod and verify the relationship between the change in resistance and the change in lengtا.


### Solution1

1. **First, we use the formula for the gauge factor (G):**

$$
G = \frac{\frac{\Delta R}{R}}{\frac{\Delta L}{L}}
$$

Where:
- \( G = 2.5 \)
- \( R = 150 R \)
- \( ΔR = 1.2 R\)
- \( L = 2 \)
- \( ΔL = 0.002 \)

2. **Calculate the ratio of the change in resistance to the initial resistance:**

$$
\frac{\Delta R}{R} = \frac{1.2}{150} = 0.008
$$

3. **Calculate the ratio of the change in length to the initial length:**

$$
\frac{\Delta L}{L} = \frac{0.002}{2} = 0.001
$$

4. **Substitute into the gauge factor formula:**

$$
2.5 = \frac{0.008}{0.001}
$$

This shows that the equation holds true and confirms the relationship between the change in resistance and the change in length.

5. **Now, we calculate the strain (ε) on the rod. Strain is given by:**

$$
\epsilon = \frac{\Delta L}{L}
$$

$$
\epsilon = \frac{0.002}{2} = 0.001
$$

6. **Conclusion:**
   - The strain on the rod is **0.001** (or 1000 microstrain).
   - This result confirms that the strain gauge accurately measures the change in length as a result of the applied force.





# Calibration of devices for measuring with strain gages
**Introduction**             
A strain gage bridge converts mechanical strain into an electrical signal that represents the measured variable. This signal goes through a series of devices—called a measurement chain—which process and display the measured value. To make sure the output correctly reflects the actual quantity, calibration is needed.
Calibration is the process of establishing a clear relationship between the output signal and the measured value. Depending on the situation, it can be simple (e.g., using known weights) or complex and expensive (e.g., applying controlled forces using hydraulic systems on large structures). Regardless of the method, accurate calibration is essential to ensure measurement reliability and minimize uncertainty.

**Measurement chain**                       
Strain gage-based measurement systems operate as measurement chains, starting from a transducer—a strain gage bridge circuit either installed directly on the object or integrated in a measurement device. This circuit converts the mechanical quantity into an analog electrical signal.
The signal then passes through one or more signal-conditioning units (e.g., amplifiers), which prepare it for the output unit. The final output can be analog, digital, printed, or fed to a computer for further processing.
The design of a measurement chain depends on several factors: the type and range of the measured variable, the required accuracy, and the desired format of the output. Some chains are permanent (e.g., in industrial applications), while others are temporarily assembled for specific tasks—especially when using self-installed strain gages.
The final device in the chain delivers the measured value in an interpretable form—ranging from simple analog meters to complex digital displays with graphical interfaces.

**Calibration of the entire measurement chain
 as a measuring device**                      
 When a measurement chain is used consistently for a specific task, it's beneficial to calibrate the entire chain as a single unit. This is done by applying known mechanical loads (e.g., force or pressure) to the transducer and comparing the displayed value with the actual input. Step-by-step loading allows mapping the relationship between input load and output signal.
During calibration, adjustments can be made to the amplifier or display to:                  
- Ensure accurate zeroing,                 
- Align the display with the correct units,                 
- Refine scale interpolation.              
Calibration involves tuning the characteristic curve of the amplifier, which defines the relationship between input and output signals. Typically, this curve is linear. Adjustments such as zero balance and gain control are used to align the system’s response with the true values of the measured variable.

![My Image](./images/photo1.png)

Calibration of a measurement chain involves adjusting zero balance (shifts the curve) and sensitivity (changes the slope). Once calibrated, the system maintains accuracy if sensitivity remains unchanged.
This method doesn’t require knowing intermediate signal values—just accurate mechanical reference inputs. However, limitations arise when known mechanical values (like exact strain or high forces) are unavailable, in which case alternative calibration methods must be used.

**Compensators**                               
Compensators work by nullifying the measurement signal using a zero-adjustment mechanism. The position of this mechanism represents the measured value.
Modern compensators perform this digitally with circuits that adjust until the input signal equals the compensation voltage. The result is displayed digitally.
When paired with a calibrated transducer, compensators offer ready-to-use measurement without user-side calibration. They are especially suited for strain gage measurements, with internal calibration based on the standard relation:
1 mV/V = 2000 µm/m (for gage factor k = 2).
A dial allows correction if k ≠ 2, eliminating the need for manual calculations.

**Shunt calibration**                       
Shunt resistors are connected in parallel to one or more arms of a strain gage bridge to create a known imbalance in the circuit. This simulates a specific mechanical strain (like compression) by artificially reducing the resistance in one bridge arm.


![My Image](./images/photo2.png)


The placement of the shunt resistor and its switching point affects calibration accuracy, especially at high carrier frequencies (e.g. 5 kHz), due to cable capacitance imbalances.
- For DC or low frequencies (e.g. 225 Hz), placement is less critical.
- For high frequencies, best accuracy is achieved by placing the shunt and switch near the transducer.
- A remote shunt in the amplifier can simplify wiring but may only generate a reference signal.

To create a specific calibration signal, the required shunt resistance can be calculated from the bridge circuit.

In a strain gauge bridge circuit, the supply voltage is \( VB = 12V \), the bridge resistance is \( R = 1000Ω \), and the output voltage is \( VM = 0.005V \). The desired output signal is 10 mV/V, meaning the target output voltage should be \( 0.01V \). The gauge factor (\( k \)) of the strain gauge is 2, and the temperature coefficient of resistance (\(α \)) for the strain gauge material is \( 0.0001 1/°C \). The task is to first calculate the required shunt resistor (\( RS \)) that will produce the desired output voltage of \( 0.01V \). Then, determine the effect of a temperature change of 30°C on the strain gauge output, considering the temperature coefficient and the other given parameters.





**Step 1: Calculate the Required Shunt Resistor \( RS \)**

We are given the following values:

- Supply voltage: \( VB = 12V \)  
- Bridge resistance: \( R = 1000Ω \)  
- Measured output voltage: \( VM = 0.005V \)  
- Desired sensitivity: \( 10\,\text{mV/V} \), so the target output voltage is \( VM = 0.01V \)

Bridge Output Approximation

In a Wheatstone bridge where only one resistor changes (i.e., one active strain gauge), the output voltage is approximately:

$$
V_M = \frac{VB}{4} \cdot \frac{\Delta R}{R}
$$

Solving for \( \Delta R \):

$$
0.01 = \frac{12}{4} \cdot \frac{\Delta R}{1000}
$$

$$
0.01 = 3 \cdot \frac{\Delta R}{1000}
$$

$$
\Delta R = \frac{0.01 \cdot 1000}{3} = 3.33\,\Omega
$$

Effective Resistance with Shunt Resistor

To produce this \( \Delta R \), the original \( 1000\Ω \) resistor must be reduced to:

$$
R_{\text{eq}} = R - \Delta R = 1000 - 3.33 = 996.67\,\Omega
$$

We need to find a resistor \( R_S \) that, when connected in parallel with \( R = 1000\,\Omega \), gives:

$$
\frac{R \cdot RS}{R + RS} = 996.67
$$

Substitute \( R = 1000 \):

$$
\frac{1000 \cdot R_S}{1000 + R_S} = 996.67
$$

Multiply both sides:

$$
1000 \cdot R_S = 996.67 (1000 + R_S)
$$

Expand:

$$
1000 R_S = 996670 + 996.67 R_S
$$

Bring like terms together:

$$
1000 R_S - 996.67 R_S = 996670
$$

$$
3.33 R_S = 996670
$$

Solve for \( R_S \):

$$
R_S = \frac{996670}{3.33} \approx 299300\,\Omega
$$

**Answer (Step 1)**

The required shunt resistor is:

$$
\boxed{R_S \approx 299.3\,\text{k}\Omega}
$$

**Step 2: Determine the Effect of a Temperature Change**

We are given:

- Temperature coefficient of resistance: \( \alpha = 0.0001\,/^{\circ}C \)  
- Temperature change: \( \Delta T = 30^{\circ}C \)  

The resistance change due to temperature is:

$$
\Delta R_T = R \cdot \alpha \cdot \Delta T = 1000 \cdot 0.0001 \cdot 30 = 3\,\Omega
$$

Now calculate the new output voltage caused by temperature:

$$
V_T = \frac{V_B}{4} \cdot \frac{\Delta R_T}{R} = \frac{12}{4} \cdot \frac{3}{1000}
$$

$$
V_T = 3 \cdot 0.003 = 0.009\,V
$$

**Answer (Step 2)**

The output voltage will increase by:

$$
\boxed{0.009\,V} \quad \text{due to a 30°C temperature rise}
$$










# Type of Strain Gauge :             
While there are many types of strain gauges--for different applications and the degree of freedom to be measured, all of them use a Wheatstone bridge to calculate the change in resistance. 
                         
**Quarter Bridge Strain Gauge**                
If you are measuring a single axis, a quarter bridge strain gauge is used, as shown in the figure below. The quarter bridge refers to that fact that only one of the four resistors is variable (Rx) and the other three resistors are fixed. The circuit determines the value of the variable resistor so that the circuit is balanced and no current passes between points B and C.

![My Image](./images/photo4.png)


**Strain Gauge Rosettes**                         
Strain gauge rosettes are sensors used to measure strain in multiple directions to determine the complete strain state on a surface, including normal, shear, and principal strains. A biaxial rosette uses two perpendicular gauges, while a triaxial rosette uses three gauges arranged at angles of 0°-45°-90° or 0°-60°-120°, depending on the required measurements.                        

![My Image](./images/photo5.png)

**Piezoresistor**                    
Piezoresistors are materials whose electrical resistance changes in response to mechanical deformation (such as stretching or compression). This property is exactly what is used to measure strain.           
Differences from conventional strain gauges:                        
  - Metal strain gauges are usually made of metal foil and their resistance change due to strain is relatively small.

 - Piezoresistors, which are often made of semiconductor materials (such as silicon), are much more sensitive to strain because their resistance changes are much larger.

Application:                       
Piezoresistors are commonly used in microelectronic sensors and MEMS systems for precise strain measurement.
