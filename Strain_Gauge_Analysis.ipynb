## What is a Strain Gauge and How Do They Work?       
Strain is a dimensionless measurement that is a ratio of the change in length to the,
original length of an object. Therefore, a positive strain is the result of stretching
a material and negative strain is the result of compression. Stress is a measurement of
the force applied divided by the initial cross-sectional area of an object, or the internal,
resisting capacity of an object....
Each strain gauge is composed of a metal foil insulated by a flexible substrate, as
shown in the figure under. The two leads pass a current through the gauge, and as the,
surface of the object being measured stretches or contracts, the change in resistance is
measured. This change in resistance is proportional to the change in length on the,
surface of the object being tested, as shown in the equation below.  In simple terms: This formula states that the change in resistance of a Strain Gauge is directly proportional to the strain of the object to which it is attached.


$$ \frac{\Delta R}{R} = k \cdot \varepsilon $$  

Where:  
- \(R\) = Resistance (**Ω**)  
- \(ΔR\) = Change in resistance (**Ω**)  
- \(k\) = Gauge Factor  
- \(ε\) = Strain

## Strain Gauge Applications:                 
The fields of civil engineering and geotechnical monitoring regularly use strain gauges
to detect failures in structures like bridges, buildings, and much more. These structures
require constant monitoring because any significant deformation could lead to injuries
or death. These gauges are commonly used because they have high precision, function
well at long distances from the test object.

 
For example, aerospace applications use millions of strain gauges to verify results from
CAD (Computer Aided Design) and FEA (Finite Element Analysis) simulations.
These tests are often conducted in dynamic conditions to display an accurate representation of how different forces affect aircraft.

## Working Principle of Strain Gauge                  
Their working principle is based on piezo-resistive effect. The piezo-resistive effect states that if
a metal conductor is stretched or compressed,its length and diameter changes. Therefore its resistance also gets changed.
Similarly the resistance of a metal conductor changes when it is strained.......

Let the change in length = ΔL and change in diameter = ΔD. To find the change in resistance (ΔR) due to applied force we use Poisson’s ratio.

</br>

Using Poisson ratio (μ) = $$
\frac{\frac{\Delta D}{D}}{\frac{\Delta L}{L}}
$$


But,

$$\frac{\Delta L}{L} = \sigma = \text{Strain}$$
Therefore,

$$
\Delta D = \mu \cdot \sigma \cdot D
$$

Thus, the change in resistance is given by:

$$
\Delta R = R \sigma (1 + 2\mu)
$$

Derivation of the formula: ΔR = R ε (1 + 2μ)

We want to derive the relationship between the change in electrical resistance of a metal wire and the applied tensile strain.

---

**Step 1: Definition of resistance**

The resistance of a wire is given by the formula:

$$
R = \rho \frac{L}{A}
$$


**Step 2: Differentiating the resistance formula**

If the wire is stretched, both \( L \) and \( A \) change. The relative change in resistance is:

$$
\frac{\Delta R}{R} = \frac{\Delta \rho}{\rho} + \frac{\Delta L}{L} - \frac{\Delta A}{A}
$$
      
We assume that the resistivity (ρ) does not change during the stretching of the wire, or that its change is very small and can be neglected.

$$
\frac{\Delta R}{R} = \frac{\Delta L}{L} - \frac{\Delta A}{A}
$$



**Step 3: Effect of Poisson’s ratio on cross-sectional area**

If the wire is stretched axially, the lateral dimensions decrease due to Poisson’s effect. For a circular cross-section:

$$
A = \pi r^2
$$

The relative change in radius due to Poisson's ratio \( \mu \) is:

$$
\frac{\Delta r}{r} = -\mu \frac{\Delta L}{L}
$$

So the relative change in area is:

$$
\frac{\Delta A}{A} = 2 \frac{\Delta r}{r} = -2\mu \frac{\Delta L}{L}
$$



**Step 4: Substitute into the resistance equation**

Now substitute \( \frac{\Δ A}{A} \) into the earlier equation:

$$
\frac{\Delta R}{R} = \frac{\Delta L}{L} - (-2\mu \frac{\Delta L}{L}) = \frac{\Delta L}{L}(1 + 2\mu)
$$

Define strain as:

$$
\varepsilon = \frac{\Delta L}{L}
$$

Then:

$$
\frac{\Delta R}{R} = \varepsilon (1 + 2\mu)
$$

Multiplying both sides by \( R \):

$$
\Delta R = R \varepsilon (1 + 2\mu)
$$



**Final Result:**

$$
\boxed{\Delta R = R \varepsilon (1 + 2\mu)}
$$

This shows how the resistance of a metallic wire increases under tensile strain due to both elongation and reduction in cross-sectional area.


From the above equation, we can conclude that the resistance of the wire increases due to the applied force. The gauge factor (G) of a strain gauge is defined as the unit change in resistance per unit change in length. Therefore,

$$
G = \frac{\frac{\Delta R}{R}}{\frac{\Delta L}{L}}
$$


(Ranges from 1.5 to 2.5)






### Question1

You are working on a project in civil engineering and need to measure the strain on a metal rod subjected to tensile force. The metal rod has an initial resistance of 150 Ω and is equipped with a strain gauge. When the rod is stretched due to the applied force, the resistance changes by 0.375 Ω. The length of the rod before stretching is 2 meters, and after stretching, it changes by 0.002 meters. The gauge factor (G) for the strain gauge is 2.5.

Calculate the strain (ε) on the rod and verify the relationship between the change in resistance and the change in lengtا.


### Solution1

1. **First, we use the formula for the gauge factor (G):**

$$
G = \frac{\frac{\Delta R}{R}}{\frac{\Delta L}{L}}
$$

Where:
- \( G = 2.5 \)
- \( R = 150 R \)
- \( ΔR = 0.375 R\)
- \( L = 2 \)
- \( ΔL = 0.002 \)

2. **Calculate the ratio of the change in resistance to the initial resistance:**

$$
\frac{\Delta R}{R} = \frac{0.375}{150} = 0.0025
$$

3. **Calculate the ratio of the change in length to the initial length:**

$$
\frac{\Delta L}{L} = \frac{0.002}{2} = 0.001
$$

4. **Substitute into the gauge factor formula:**

$$
2.5 = \frac{0.0025}{0.001}
$$

This shows that the equation holds true and confirms the relationship between the change in resistance and the change in length.

5. **Now, we calculate the strain (ε) on the rod. Strain is given by:**

$$
\epsilon = \frac{\Delta L}{L}
$$

$$
\epsilon = \frac{0.002}{2} = 0.001
$$

6. **Conclusion:**
   - The strain on the rod is **0.001** (or 1000 microstrain).
   - This result confirms that the strain gauge accurately measures the change in length as a result of the applied force.





# Calibration of devices for measuring with strain gages
**Introduction**             
A strain gage bridge converts mechanical strain into an electrical signal that represents the measured variable. This signal goes through a series of devices—called a measurement chain—which process and display the measured value. To make sure the output correctly reflects the actual quantity, calibration is needed.
Calibration is the process of establishing a clear relationship between the output signal and the measured value. Depending on the situation, it can be simple (e.g., using known weights) or complex and expensive (e.g., applying controlled forces using hydraulic systems on large structures). Regardless of the method, accurate calibration is essential to ensure measurement reliability and minimize uncertainty.

**Measurement chain**                       
Strain gage-based measurement systems operate as measurement chains, starting from a transducer—a strain gage bridge circuit either installed directly on the object. This circuit converts the mechanical quantity into an analog electrical signal.
The signal then passes through one or more signal-conditioning units (e.g., amplifiers), which prepare it for the output unit. The final output can be analog, digital, printed, or fed to a computer for further processing.
The design of a measurement chain depends on several factors: the type and range of the measured variable, the required accuracy, and the desired format of the output. Some chains are permanent (e.g., in industrial applications), while others are temporarily assembled for specific tasks.





**Shunt Calibration**:
Shunt calibration uses resistors connected in parallel to a strain gage bridge arm to simulate mechanical strain by creating a known imbalance in the circuit.

## Type of Strain Gauge :   
The two types of strain gauges as  :


- Bonded Strain Gauge           
- Unbonded Strain Gauge

**Bonded Strain Gauge**    

                                

![My Image](./images/photo4.png)           

                   
A bonded strain gauge consists of a fine wire grid attached to a thin paper sheet and bonded to the surface under test. When force is applied, the wire’s length and diameter change, altering its resistance. This change is measured using a Wheatstone bridge. Bonded strain gauges are suitable for measuring small displacements, and for larger displacements, they are mounted on flexible elements like cantilever beams.

**Unbonded Strain Gauge**   

                          

![My Image](./images/photo5.png)       

                          
An unbonded strain gauge consists of a wire mounted on a fixed frame with a movable central armature. Movement of the armature stretches the wires, changing their resistance. This change is measured using a Wheatstone bridge, and the output voltage can be calibrated based on the armature’s displacement.
